# EDA and Data cleaning

```python
airbnb.drop_duplicates(inplace=True)
```

```python
airbnb.shape
```

541 duplicated rows have been removed

```python
# Dropping columns with missing values greater than 50%
print(airbnb.columns[airbnb.isnull().mean() > 0.5], "\n")
airbnb = airbnb.drop(airbnb.columns[airbnb.isnull().mean() > 0.5], axis=1)
```

```python
# Convert 'last review' column to datetime type
airbnb['last review'] = pd.to_datetime(airbnb['last review'])
```

```python
airbnb['price'] = airbnb['price'].str.replace('$', '').str.replace(',', '').astype(float)
airbnb['service fee'] = airbnb['service fee'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
airbnb['price']
```

```python
airbnb['service fee']
```

Dropping the null value rows from all the above mentioned features

```python
Drop_list = ['price','NAME','host_identity_verified', 'neighbourhood group','neighbourhood', 'cancellation_policy', 'instant_bookable', 'host name','Construction year', 'lat','long']

for i in Drop_list:
    airbnb = airbnb.dropna(subset = [i])
```

```python
airbnb.shape
```

Droping columns where value counts is equal to 1

```python
for i in airbnb.columns:
    if airbnb[i].value_counts().shape[0] == 1:
        airbnb = airbnb.drop(i, axis=1)
```

Fill null values in "service fee" by zeros 

```python
airbnb['service fee'] = airbnb['service fee'].fillna(0)
```

remove availability 365 that less than 0 and greater than 365

```python
airbnb['availability 365'] = np.where(airbnb['availability 365']<0, airbnb['availability 365']*-1, airbnb['availability 365'])
airbnb['availability 365'] = np.where(airbnb['availability 365']>365, 365, airbnb['availability 365'])
airbnb['minimum nights'] = np.where(airbnb['minimum nights']<0, airbnb['minimum nights']*-1, airbnb['minimum nights'])
```

## Encoding and splitting the data

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score 
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
```

```python
airbnb["reviews per month"].fillna(airbnb["reviews per month"].mean(),inplace=True)
airbnb["minimum nights"].fillna(airbnb["minimum nights"].mean(),inplace=True)
airbnb["availability 365"].fillna(airbnb["availability 365"].mean(),inplace=True)
airbnb["calculated host listings count"].fillna(airbnb["calculated host listings count"].mean(),inplace=True)
airbnb["number of reviews"].fillna(airbnb["number of reviews"].median(), inplace=True)
airbnb["review rate number"].fillna(airbnb["review rate number"].median(), inplace=True)
```

```python
airbnb.columns
```

```python
airbnb.dtypes
```

```python
airbnb['host_identity_verified'].value_counts()
```

```python
le = LabelEncoder()
airbnb['host_identity_verified'] = le.fit_transform(airbnb['host_identity_verified'])
```

```python
airbnb['instant_bookable'].value_counts()
```

```python
airbnb['instant_bookable'] = le.fit_transform(airbnb['instant_bookable'])
airbnb['cancellation_policy'] = le.fit_transform(airbnb['cancellation_policy'])
airbnb['room type'] = le.fit_transform(airbnb['room type'])
airbnb['neighbourhood group'] = le.fit_transform(airbnb['neighbourhood group'])
```

```python
airbnb.dtypes
```

```python
airbnb_split_data = airbnb[['host_identity_verified','neighbourhood group','instant_bookable','cancellation_policy',
                            'room type','Construction year','minimum nights','number of reviews', 'reviews per month',
                            'review rate number', 'calculated host listings count','availability 365','price',
                            'service fee']]
```

```python
x =airbnb_split_data.drop(["price"],axis=1).values
y= airbnb_split_data['price'].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
airbnb_split_data.isnull().sum().sort_values()
```

## Modelling

### 1) Linear Regression

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(x_train,y_train)
lr_score =lr_model.score(x_train,y_train)
lr_score
```

```python
lr_model.score(x_test,y_test)
```

```python
y_pred = lr_model.predict(x_test)
mse = mean_squared_error(y_test , y_pred)
mse
```

### 2) Decission Tree 

```python
DT_model = DecisionTreeRegressor().fit(x_train, y_train)
DT_model_v = DT_model.score(x_train,y_train)
DT_model_v 
```

```python
DT_model.score(x_test,y_test)
```

```python
y_pred = DT_model.predict(x_test)
```

```python
DT_score = r2_score(y_test , y_pred)
DT_score
```

```python
mse = mean_squared_error(y_test , y_pred)
mse
```

### 3) RandomForest Regressor 

```python
RF_model = RandomForestRegressor().fit(x_train, y_train)
RF_model_v = RF_model.score(x_train,y_train)
RF_model_v
```

```python
print(RF_model.score(x_train , y_train))
print(RF_model.score(x_test , y_test))
```

```python
y_pred = RF_model.predict(x_test)
```

```python
rf_score = r2_score(y_test , y_pred)
rf_score
```

### 4) Ridge Regression 

```python
RR_model = linear_model.Ridge(alpha=0.4)
RR_model.fit(x_train , y_train)
RR_model_v = RR_model.score(x_train , y_train)
RR_model_v
```

```python
print(RR_model.score(x_train , y_train))
print(RR_model.score(x_test , y_test))
```

```python
y_pred = RR_model.predict(x_test)
rid_score = r2_score(y_test , y_pred)
rid_score
```

```python
mse = mean_squared_error(y_test , y_pred)
mse
```

### Comparing models

```python
dic = {'linear Regression':lr_score,'Decision Tree':DT_model_v,'RandomForest Regressor':RF_model_v,'Ridge Regression':RR_model_v}
```

```python
models = list(dic.keys())
values = list(dic.values())
plt.rc('font', size=20)

plt.figure(figsize=(30,10))
plt.xlabel('Model name', fontsize=25)
plt.ylabel('Accuracy', fontsize=25)

plt.bar(range(len(dic)), values, tick_label=models,color='lightblue')
plt.show()
```
